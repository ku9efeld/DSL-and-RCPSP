```python
GEN = 4
select_k = 14
n = 7
res = [select_k  if n // 2 + n - select_k >= 0 else  n // 2 + n]
gen_cand = [n // 2]
for i in range(1, GEN):
    res.append( select_k if res[i-1] // 2 + res[i-1] - select_k > 0 else res[i-1] // 2 + res[i-1] )
    gen_cand.append(res[i-1] // 2)
```
